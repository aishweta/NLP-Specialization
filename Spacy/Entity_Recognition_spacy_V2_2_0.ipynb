## 1. Part-of-Speech (POS) Tagging using spaCy
### What is POS Tagging?
 It is a process of converting a sentence to forms â€“ list of words, list of tuples (where each tuple is having a form (word, tag)). The tag in case of is a part-of-speech tag, and signifies whether the word is a noun, adjective, verb, and so on.

image.png

### Why POS tag is used
Some words can function in more than one way when used in different circumstances. 
- The POS Tagging here plays a crucial role to understand in what context the word is used in the sentence. 
- POS Tagging is useful in sentence parsing, information retrieval, sentiment analysis, etc.

```python
doc1 = nlp1("im learning awesome python and machine learning")
```

```python
for token in doc1:
    # Print the token and its part-of-speech tag
    print(token.text, "-->", token.pos_)
```

## 2. Lemmatization:
- Stemming and lemmatization are the techniques used to reducing a word into root word.

e.g.==> Running= run, waiting  ==> wait,etc

- But stemming have its own drawbacks . like Stemming just removes or stems the last few characters of a word.

e.g. Studies => Studi

We can see that using stemming those  words are also generated who have no meaning.

#### In spacy only Lemmatization is present.

```python
doc2=nlp1("Rajesh went to school. But today there is raining outside. But he is good in studies ")
for token in doc2:
    print(token, token.lemma_)
```

So we can see that studies is converted to study. So we can say that lemmatization returns meaningful words.

```python
### dataset format used for spacy 2.0.0+ version and its model building

# load spacy model
nlp=spacy.load('en_core_web_sm')

# Initiate class: Getting the pipeline component
ner=nlp.get_pipe("ner")
```

("Walmart is a leading e-commerce company and amazon is competitor", {"entities": [(0, 7, "ORG"), (23,29 "ORG")]})


```python
TRAIN_DATA = [
              ("Walmart is a leading e-commerce company", {"entities": [(0, 7, "ORG")]}), # (0,n-1)
              ("I reached Chennai yesterday.", {"entities": [(19, 28, "GPE")]}),
              ("I recently ordered a book from Amazon", {"entities": [(24,32, "ORG")]}),
              ("I was driving a BMW", {"entities": [(16,19, "PRODUCT")]}),
              ("I ordered this from ShopClues", {"entities": [(20,29, "ORG")]}),
              ("Fridge can be ordered in Amazon ", {"entities": [(0,6, "PRODUCT")]}),
              ("I bought a new Washer", {"entities": [(16,22, "PRODUCT")]}),
              ("I bought a old table", {"entities": [(16,21, "PRODUCT")]}),
              ("I bought a fancy dress", {"entities": [(18,23, "PRODUCT")]}),
              ("I rented a camera", {"entities": [(12,18, "PRODUCT")]}),
              ("I rented a tent for our trip", {"entities": [(12,16, "PRODUCT")]}),
              ("I rented a screwdriver from our neighbour", {"entities": [(12,22, "PRODUCT")]}),
              ("I repaired my computer", {"entities": [(15,23, "PRODUCT")]}),
              ("I got my clock fixed", {"entities": [(16,21, "PRODUCT")]}),
              ("I got my truck fixed", {"entities": [(16,21, "PRODUCT")]}),
              ("Flipkart started it's journey from zero", {"entities": [(0,8, "ORG")]}),
              ("I recently ordered from Max", {"entities": [(24,27, "ORG")]}),
              ("Flipkart is recognized as leader in market",{"entities": [(0,8, "ORG")]}),
              ("I recently ordered from Swiggy", {"entities": [(24,29, "ORG")]})
              ]
```

```python
for _, annotations in TRAIN_DATA:
    for ent in annotations.get("entities"):
        ner.add_label(ent[2])
```

```python
pipe_exceptions = ["ner", "trf_wordpiecer", "trf_tok2vec"]
unaffected_pipes = [pipe for pipe in nlp.pipe_names if pipe not in pipe_exceptions]
```

```python
# Import requirements
import random
from spacy.util import minibatch, compounding
from pathlib import Path

# TRAINING THE MODEL
with nlp.disable_pipes(*unaffected_pipes):

  # Training for 30 iterations
  for iteration in range(30):

    # shuufling examples  before every iteration
    random.shuffle(TRAIN_DATA)
    losses = {}
    # batch up the examples using spaCy's minibatch
    batches = minibatch(TRAIN_DATA, size=compounding(4.0, 32.0, 1.001))
    for batch in batches:
        texts, annotations = zip(*batch)
        nlp.update(
                    texts,  # batch of texts
                    annotations,  # batch of annotations
                    drop=0.5,  # dropout - make it harder to memorise data
                    losses=losses,
                )
        print("Losses", losses)
```

```python
# Testing the model
doc = nlp("I like to driving a BMW")
print("Entities", [(ent.text, ent.label_) for ent in doc.ents])
##  Visually we can observe how entities looks
displacy.render(doc,style="ent",jupyter=True)
```

```python
# Testing the model
doc = nlp("I love to drive a BMW, but i drive TVS")
print("Entities", [(ent.text, ent.label_) for ent in doc.ents])
##  Visually we can observe how entities looks
displacy.render(doc,style="ent",jupyter=True)
```
