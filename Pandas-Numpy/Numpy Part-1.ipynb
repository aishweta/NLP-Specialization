```python
## create one dimensional array with dtype
array1 = np.array([1,2,3,4], dtype='int32')
print(array1)
print(array1.dtype)
print(array1.itemsize)
print(array1.nbytes)
```

```python
# create 3 dimentional array
array33 = np.array( [[[1,2], [3,4]]] )
print(array33)
print(array33.ndim)
print
```

```python
## create one dimension to multiple dimensions
array3 = np.array([1,2,3,4,5,6], ndmin = 4) 
print(array3)
```

```python
## convert into dtype parameter
array4 = np.array([1, 2, 3, 4], dtype = complex) 
print(array4)
```
